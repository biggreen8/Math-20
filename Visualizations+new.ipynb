```python
import re
import pickle
import itertools
import matplotlib.pyplot as plt
import os
```

```python
# Directory where the .pkl files are saved (update this if needed)
directory = "/Users/ethansun/Downloads/Math 20 stuff"  # Replace with the actual path
```

```python
# Define the pattern for extracting the relevant addition problems
ptrn = r"(\d+)\s+\+\s+(\d+)\s=\s(\d+)"
search = lambda x: re.search(ptrn, x.strip())
```

```python
# Iterate over each .pkl file and generate a bar graph
for i in range(1, 7):
    accuracies = []
    x = list(range(0, i+1))
    
    for j in range(0, i+1):
        # Construct the file path
        filename = f"{i}digit-{j}carry-results+new.pkl"
        filepath = os.path.join(directory, filename)

        # Load the .pkl file
        with open(filepath, "rb") as f:
            carries = pickle.load(f)

        # Flatten the list (if needed) and convert bytes to strings
        if isinstance(carries, list):
            carries = list(itertools.chain.from_iterable(carries))
        
        # Debug: Check if carries are being loaded correctly
        print(f"Loaded {len(carries)} entries from {filename}")

        # Filter for only lines matching the addition pattern
        filtered = []
        for v in carries:
            # Decode to string if it's a bytes object
            if isinstance(v, bytes):
                v = v.decode("utf-8", errors="ignore")
            
            match = search(v)
            if match:
                num1, num2, result = int(match.group(1)), int(match.group(2)), int(match.group(3))
                filtered.append((num1, num2, result))
                
        # Debug: Check if filtered results are being computed correctly
        print(f"Filtered results for carry {j}: {filtered}")
        print("Length of filtered = " + str(len(filtered)))

        # Calculate accuracy for this specific carry value (j)
        for u in filtered:
            if u[0] + u[1] == u[2]:
                print(str(u[0]) + " + " + str(u[1]) + " = " + str(u[2]) + ": True")
            else:
                print(str(u[0]) + " + " + str(u[1]) + " = " + str(u[2]) + ": False")
        acc = len([v for v in filtered if v[0] + v[1] == v[2]]) / len(filtered)

        print(f"Calculated accuracy for {j}-th digit carry: {acc}")
        accuracies.append(acc)

    # Create the bar graph for this particular digit carry scenario
    plt.bar(x, accuracies)
    plt.ylim(0, 1)
    plt.xlabel("Digit Carried")
    plt.ylabel("Accuracy")
    plt.title(f"Accuracy for {i}-digit addition problems")
    plt.xticks(x)  # Ensure x-axis shows integers 0 to i
    plt.savefig(f"{i}-digit-addition-accuracy-results.pdf")
    plt.clf()  # Clear the figure for the next plot
```
